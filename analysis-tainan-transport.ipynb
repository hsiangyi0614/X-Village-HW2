### 1.如何取得資料?取什麼資料?
我使用台南市政府資料開放平臺取得我要分析的資料。
我要分析的是'大台南公車運量資訊（103~106年）'data。
### 2.分析用途 :
目標一開始訂定在透過目前的大眾運輸系統上(公車)去評估是否蓋台南捷運之探討。
### 3.分析資料 :
將JSON讀進後看檔案的屬性，每個LIST中的KEY欄位的命名代表著什麼意思。
### 4.分析的目標轉換
因為檔案讀進後與預期目標產生錯誤，首先交通局公共運輸處僅提供每條線路的月數據資料。無法精準地去比對結果可能導致與興建捷運相關性降低，但可以去比對出台南民眾對於大眾運輸系統的使用程度及依賴性。
### 5.資料視覺化
資料看過後將資料彙整後畫出該線路的成長比折線圖。
### 6.圖表及數據的呈現結果

### 7.探討公車運量數據的變化
從幹線公車中，在起初還未重新處理所有台南縣市的公車時的運量西當低迷，經由台南市府交通局透過公共運輸的推動城市進步與發展，為大台南地區推動「捷運化公共運輸系統」在102年開始整合，以每個月的方式開出了6大幹線，並使用電子票證可以由交通部補助票價26元及轉乘優惠。並於103年開始做出有效的統計。
103年度
在六大幹線中103-104年度的成長率部分線路是有相當突出，唯獨棕幹線成長高達46.99%以及黃幹線有最高來到61.29%的成長，而黃幹線103年度變動率高達118.5%，從這個數據中得知這是不太準的依據，從圖表可以得知當中3月的運量來到相當高的新高，這個值應該做利外處理。
104年度
在六大幹線中104-105年度的成長率所有路線都是平穩的成長，都落於13.5~21.6%中，變動率都較前年年度來有微幅降低，表示著可能長期使用者較為穩定。唯獨黃幹線也是相當的突出的數據表現。
105年度
在六大幹線中105-106年度的成長率有五條幹線都是負成長的，為什麼會產生負成長這個可以作為日後繼續探討的問題之 一，是乘車品質、班距過長、運量增加班次未調整等因素。紅幹線的為正成長更可以找出相關的因素，需要再透過更多的資料輔佐。在變動率的部分能有不穩定的結果產生。
106年度
從過去數據中可以看到紅幹線的變動是持續下降的，有唯獨它在幹線中都是呈現正成長的，可見這條路線以這些數據判斷覺得相當好的範例以利其他路線去學習。

```python
#將整個檔案印出
import json
import requests
url = 'http://61.60.30.138/opendata/0d56af62-d124-4d85-b4c0-12160eb07e8e?format=json'
data = requests.get(url)
data.encoding = 'utf-8'
data = json.loads(data.text)
print(data)
```
